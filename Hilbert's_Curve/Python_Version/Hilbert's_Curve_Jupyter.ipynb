```python
# Group 5
# Benzil Emmanuel: SCT211-0463/2023
# Francis Mung'ang'u: SCT211-0032/2023

# Importing Python's Built in Drawing Library
import turtle

# ---------------- Hilbert Generator ----------------
def generate_hilbert(points, x, y, xi, xj, yi, yj, order):
    if order == 0: # Best case
        px = x + (xi + yi) / 2
        py = y + (xj + yj) / 2
        points.append((px, py))
    else:
        generate_hilbert(points, x, y, yi/2, yj/2, xi/2, xj/2, order - 1)
        generate_hilbert(points, x + xi/2, y + xj/2, xi/2, xj/2, yi/2, yj/2, order - 1)
        generate_hilbert(points, x + xi/2 + yi/2, y + xj/2 + yj/2, xi/2, xj/2, yi/2, yj/2, order - 1)
        generate_hilbert(points, x + xi/2 + yi, y + xj/2 + yj,
                         -yi/2, -yj/2, -xi/2, -xj/2, order - 1)

# ---------------- Generate Points ----------------
points = []
order = 5

generate_hilbert(points,
                 -1.0, -1.0,
                  2.0,  0.0,
                  0.0,  2.0,
                 order)

# ---------------- Turtle Setup and Drawing with Turtle ----------------
t = turtle.Turtle()
t.speed(0)                        # fastest drawing
turtle.bgcolor("black")           # background
t.color("cyan")                   # pen color
t.width(2)                        # line thickness

scale = 200                       # zoom factor

#----------------- Drawing ----------------------
t.penup()
first_x, first_y = points[0]
t.goto(first_x * scale, first_y * scale) # move to first point
t.pendown()

for x, y in points:
    t.goto(x * scale, y * scale)         # draw line to each point

turtle.done()
```
